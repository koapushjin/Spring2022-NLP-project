
# Stacked bar chart

This is an example of creating a stacked bar plot with error bars
using `~matplotlib.pyplot.bar`.  Note the parameters *yerr* used for
error bars, and *bottom* to stack the women's bars on top of the men's
bars.


```python
import matplotlib.pyplot as plt


labels = ['A', 'B', 'C', 'D', 'E', 'F']
zeroshot = [0.529412, 0.529412, 0.529412, 0.527132, 0.538462, 0.586207]
prompted5 = [0.679835, 0.538832, 0.475234, 0.371252, 0.4819422, 0.639234]
fivecommon = [0.534234, 0.45892, 0.523453, 0.412213, 0.452453, 0.723234]
allcommon = [0.579234, 0.523423, 0.521344, 0.342342, 0.610383, 0.699231]
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, zeroshot, width, label='Zero-shot')
ax.bar(labels, prompted5, width, bottom=zeroshot,
       label='5 Prompts')


ax.set_ylabel('Accuracy')
ax.set_title('Prompt methods Comparison')
ax.legend()

fig.savefig('plot2.png', dpi=300, bbox_inches="tight")
plt.show()
```
