### pre-process prompts

```python
def shuffle_prompts(p):
  random.shuffle(prompts)
  prompt1="".join(prompts)
  return prompt1



def n_finds(p,c):
  s=0
  index = p.find(c,s,len(p)-1)
  Nos=[]
  while index>0:
    Nos.append(index)
    s = index+1
    index = p.find(c,s,len(p)-1)
  return len(Nos)

def generate_prompt(prompts):
  p = shuffle_prompts(prompts)
  n_Nos = n_finds(p,"No")
  n_Yes = n_finds(p,"Yes")

  if n_Nos>0:
    p.replace("No","N/A",n_Nos//2)
  if n_Yes>0:
    p.replace("Yes","N/A",n_Nos//2)
  return p
```
